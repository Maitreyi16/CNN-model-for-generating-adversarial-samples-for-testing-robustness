```python
# Simulating Epidemic Spread using SIR Model

import random

# Constants
POPULATION_SIZE = 1000
INITIAL_INFECTED = 1
INFECTION_RATE = 0.3
RECOVERY_RATE = 0.1
SIMULATION_DAYS = 100

# Initialize Population
population = ['S'] * (POPULATION_SIZE - INITIAL_INFECTED) + ['I'] * INITIAL_INFECTED
random.shuffle(population)

# Simulation Loop
for day in range(SIMULATION_DAYS):
    new_infections = 0
    recoveries = 0

    # Iterate over the population
    for i in range(len(population)):
        if population[i] == 'I':
            # Infected person interacts with others
            for j in range(len(population)):
                if population[j] == 'S' and random.random() < INFECTION_RATE:
                    population[j] = 'I'
                    new_infections += 1

            # Check for recovery
            if random.random() < RECOVERY_RATE:
                population[i] = 'R'
                recoveries += 1

    # Print current stats
    print(f'Day {day+1}: New Infections = {new_infections}, Recoveries = {recoveries}')

# Final Stats
final_infected = population.count('I')
final_recovered = population.count('R')
final_susceptible = population.count('S')

print(f'\nSimulation Complete!')
print(f'Final Stats:')
print(f'Susceptible: {final_susceptible}')
print(f'Infected: {final_infected}')
print(f'Recovered: {final_recovered}')
```
